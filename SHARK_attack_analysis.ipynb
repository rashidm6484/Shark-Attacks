```python
#import File
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('attacks.csv', encoding='latin1')
```

```python
# Drop Unnecessary Columns
df = df.drop(columns=['Case Number.1', 'Case Number.2', 'href formula', 'href', 'Fatal (Y/N)','pdf', 'original order'], errors='ignore')
```

```python
# Cleaning Column
df.columns = df.columns.str.strip()

# Apply general replacements for other column names
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('(', '')
df.columns = df.columns.str.replace(')', '')
```

```python
# Remove rows where Case Number is missing
df = df[df['Case_Number'].notna()]
```

```python
#Remove duplicate rows
df = df.drop_duplicates()
```

```python
# For misssing value
df = df.replace(' ', np.nan)

# Check missing values
df.isnull().sum()
```

### **Cleaning and Standardizing Gender Column**

```python
df = df.rename(columns={'Sex': 'Gender'})
# Clean and standardize 'Gender' column values
df['Gender'] = df['Gender'].astype(str).str.strip().str.upper()

# Identify invalid gender entries and replace with NaN
print(df['Gender'].unique())
```

```python
# Based on common invalid entries observed in similar datasets
invalid_gender_values = ['.', 'N', 'UNOC', 'LLI', '2017', 'UNKNOWN', 'NAN'] # Added 'NAN'
df['Gender'] = df['Gender'].replace(invalid_gender_values, np.nan)


# Drop rows where 'Gender' is NaN after cleaning
df.dropna(subset=['Gender'], inplace=True)
```

```python
# Further standardize to 'M' and 'F' if other variations exist
df['Gender'] = df['Gender'].replace({'FEMALE': 'F', 'MALE': 'M'})

# Display unique values and counts to confirm cleaning
print("Unique values and counts in 'Gender' column after cleaning:")
print(df['Gender'].value_counts())
```

```python
# Extract hour from 'Time' and convert to numeric, handling errors
df['Hour'] = pd.to_numeric(df['Time'].str.extract('^(\d{2})')[0], errors='coerce')

# Categorize hours into time of day
def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    elif 21 <= hour <= 23 or 0 <= hour < 5:
        return 'Night'
    return np.nan

# Apply the function to create the 'Time_of_Day' column
df['Time_of_Day'] = df['Hour'].apply(get_time_of_day)
```

```python
#Cleaning 'Year' Column

# Identify and remove invalid year entries (e.g., years before 1917 or future years)
df = df[df['Year'] >= 1917] # Set minimum year to 100 years prior to 2017
df = df[df['Year'] <= 2023] # Assuming the data should not exceed the current year

# Convert Year to integer type
df['Year'] = df['Year'].astype(int)
```

```python
#clean date column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows where 'Date' is NaN after cleaning
df.dropna(subset=['Date'], inplace=True)
```

```python
# Convert 'Age' to numeric, coercing errors to NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
# Identify unique values in 'Type' column before cleaning
print(df['Type'].unique())

# Calculate the mode of the 'Type' column
mode_type = df['Type'].mode()[0]
```

```python
# Replace NaN values with the mode
df['Type'].fillna(mode_type, inplace=True)
# Replace 'Invalid' and any other specific invalid strings (e.g., typos) with the mode
df['Type'] = df['Type'].replace(['Invalid', 'UNKNOWN', 'unknown', 'nan', ' '], mode_type)
```

```python
# Final check
df.info()
df.head()
```

# **Exploratory Data Analysis**

---



# **1.Top 10 Countries with Shark Attacks**

```python
df['Country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries with Shark Attacks')
plt.show()
```

*This bar chart highlights the top countries with the highest number of shark attacks.with the visualisation higest attakck held on usa*

# **2. Top 10 Activities Leading to Attacks by type**

```python
# Get the top 10 activities
top_activities = df['Activity'].value_counts().head(10).index

# Filter the DataFrame to include only these top activities
df_top_activities = df[df['Activity'].isin(top_activities)]

# Create a cross-tabulation of Activity and Type
attacks_by_activity_type = pd.crosstab(df_top_activities['Activity'], df_top_activities['Type'])


attacks_by_activity_type.plot(kind='barh', stacked=True,)
plt.title('Top 10 Activities Leading to Shark Attacks by Type')
plt.xlabel('Number of Attacks')
plt.ylabel('Activity')
plt.tight_layout()
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

*This bar chart shows the most common activities people were engaged in when shark attacks occurred, identifying activities with higher risk.it seems surfing and swimming is the higest two values*

# **3. Top 10 Areas with Shark Attacks in the USA**

```python
usa_attacks = df[df['Country'] == 'USA']
top_10_areas = usa_attacks['Area'].value_counts().head(10)

colors = sns.color_palette('viridis', len(top_10_areas))
plt.pie(top_10_areas, labels=top_10_areas.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Proportion of Shark Attacks by Top 10 Areas in the USA')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

*Focusing on the USA, this bar chart highlights the specific areas (e.g., states or regions) within the country that experience the most shark attacks.
in USA more than half of the attacks held in Florida*

# **4. Shark Attacks by Time of Day**

```python
sns.countplot(y='Time_of_Day', data=df, order=['Morning', 'Afternoon', 'Evening', 'Night'], palette='viridis')
plt.title('Shark Attacks by Time of Day')
plt.xlabel('Number of Attacks')
plt.ylabel('Time of Day')
plt.tight_layout()
plt.show()
```

*This count plot analyzes the frequency of shark attacks based on the time of day, revealing if certain periods are more prone to incidents.most of the insidents in day time*

# **5. Top 10 Shark Species Involved in Attacks by Type**

```python
# Get the top 10 shark species
top_species = df['Species'].value_counts().head(10).index
# Filter the DataFrame to include only these top species
df_top_species = df[df['Species'].isin(top_species)]
# Create a cross-tabulation of Species and Type
attacks_by_species_type = pd.crosstab(df_top_species['Species'], df_top_species['Type'])


attacks_by_species_type.plot(kind='bar', stacked=True,)
plt.title('Top 10 Shark Species Involved in Attacks by Type')
plt.xlabel('Shark Species')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

*This stacked bar chart shows the top 10 shark species involved in attacks, categorized by attack type. It highlights how different types of incidents are distributed across species, making it easy to compare patterns and identify the most common attack types for each species.*

# **6.Age Distribution of Victims**

```python
sns.histplot(df['Age'].dropna(), bins=50, kde=True)
plt.title("Age Distribution of Victims")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

*The age distribution of shark attack victims shows how incidents are spread across different age groups. The visualization typically reveals that young adults (around 15â35 years) are most frequently affected, likely due to higher participation in water activities such as swimming and surfing.*

# **7.Gender-wise Shark Attacks**

```python
# Calculate gender counts
gender_counts = df['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']

# Create labels with gender and count
pie_labels = [f'{gender} ({count})' for gender, count in gender_counts.values]

plt.pie(gender_counts['Count'], labels=pie_labels, autopct='%1.1f%%')
plt.title('Gender-wise Shark Attacks')
plt.show()
```

*This visualization shows the distribution of shark attacks between males and females. It typically reveals that males are involved in a significantly higher number of attacks compared to females.*

## **8. Shark Attacks Over the Years by Gender**

```python
# Calculate the count of attacks for each year and gender
attacks_by_year_gender = df.groupby(['Year', 'Gender']).size().reset_index(name='Count')

sns.lineplot(x='Year', y='Count', hue='Gender', data=attacks_by_year_gender, marker='o', palette={'M': 'blue', 'F': 'red'})
plt.title('Shark Attacks Over the Years by Gender')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

*Each line represents a gender category, allowing for a clear comparison of how shark attack frequencies have changed over the years for males and females. The markers highlight individual yearly data points, making it easier to observe fluctuations and trends.*

*The chart reveals that shark attacks involving males are significantly higher than those involving females across most years.*

# **9.Type of Shark Attacks by Top 10 Countries**

```python
# Top 10 countries
top_countries = df['Country'].value_counts().head(10).index
df_top = df[df['Country'].isin(top_countries)]

# Create pivot table
pivot = pd.crosstab(df_top['Country'], df_top['Type'])
plt.figure()
sns.heatmap(pivot, annot=True, fmt='d')
plt.title("Type of Shark Attacks by Top 10 Countries")
plt.xlabel("Attack Type")
plt.ylabel("Country")
plt.show()
```

*This heatmap visualizes the distribution of different types of shark attacks across the top 10 countries
This visualization allows for quick comparison of attack patterns across countries, helping to identify which regions experience more frequent or specific types of shark attacks. It also highlights any dominant attack types within particular countries, making it easier to detect trends, anomalies, or regional differences.*

# **10. Shark Attacks by Activity (Gender-wise)**

```python
#top 10 activities
top_activities = df['Activity'].value_counts().head(10).index
df_top = df[df['Activity'].isin(top_activities)]

activity_gender = pd.crosstab(df_top['Activity'], df_top['Gender'])
activity_gender.plot(kind='barh', stacked=True)
plt.title("Shark Attacks by Activity (Gender-wise)")
plt.xlabel("Number of Attacks")
plt.ylabel("Activity")
plt.xticks(rotation=45)

plt.show()
```

*This visualization shows how shark attacks are distributed across different activities (such as surfing, swimming, fishing, etc.) for each gender. It helps identify which activities are most commonly associated with attacks among males and females.*

# **11. Shark Attacks Over the Years by Top 3 Countries**

```python
# Identify the top 3 countries for clearer visualization
top_10_countries = df['Country'].value_counts().head(3).index

# Filter the DataFrame to include only these top countries
df_top_10_countries = df[df['Country'].isin(top_10_countries)]

# Group by Year and Country, then count the occurrences
attacks_by_year_country = df_top_10_countries.groupby(['Year', 'Country']).size().reset_index(name='Count')

plt.figure(figsize=(14, 7))
sns.lineplot(x='Year', y='Count', hue='Country', data=attacks_by_year_country, marker='o')
plt.title('Shark Attacks Over the Years by Top 3 Countries')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

*This line plot illustrates the trend of shark attacks over the years, differentiated by the top 10 most active countries. It helps in understanding which countries have consistently high numbers of attacks and if there are any increasing or decreasing trends for specific regions over time.*
