### **Shark Attack Data Analysis**

---




## Objective

The goal of this project is to perform exploratory data analysis (EDA) on global shark attack data.
We aim to identify patterns based on year, location, activity, gender, and fatality.

## Dataset Overview

- Dataset contains shark attack records over the past 100 years
- Includes:
  - Year
  - Location
  - Activity
  - Victim details (age, gender)
  - Shark species
  - Fatality (Yes/No)

### Initial Observations:
- Contains missing values
- Mixed data types
- Some irrelevant columns

DATA SET LINK:  https://mavenanalytics.io/data-playground/shark-attacks



```python
#import File
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('attacks.csv', encoding='latin1')
```

**DATA CLEANING**

---

```python
# View first 10 rows
df.head(10)
```

```python
#DATA FRAME INFORMATIONS
df.info()

df.shape
```

```python
# Remove rows where Case Number is missing
df = df[df['Case Number'].notna()]
```

```python
# Drop Unnecessary Columns
df = df.drop(columns=['Case Number.1', 'Case Number.2', 'href formula'], errors='ignore')
```

```python
# Cleaning Column
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('(', '')
df.columns = df.columns.str.replace(')', '')
```

```python
# Convert Year Column to Numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
```

```python
#Standardize Column
df['Fatal_Y/N'] = df['Fatal_Y/N'].str.upper()
```

```python
# For misssing value
df = df.replace(' ', np.nan)

# Check missing values
df.isnull().sum()
```

```python
#Remove duplicate rows
df = df.drop_duplicates()
```

```python
# Extract hour from 'Time' and convert to numeric, handling errors
df['Hour'] = pd.to_numeric(df['Time'].str.extract('^(\d{2})')[0], errors='coerce')

# Categorize hours into time of day
def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    elif 21 <= hour <= 23 or 0 <= hour < 5:
        return 'Night'
    return np.nan
```

```python
# Convert Year to integer type
df['Year'] = df['Year'].astype(int)
```

```python
# Final check
df.info()
df.head()
```

```python
df.to_csv("attacks_cleaned.csv", index=False)
```

# **Exploratory Data Analysis**

---



**1.Top 10 Countries with Shark Attacks**

```python
df['Country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries with Shark Attacks')
plt.show()
```

*This bar chart highlights the top countries with the highest number of shark attacks.*

### 2. Top 10 Activities Leading to Attacks

```python
plt.figure(figsize=(12, 6))
df['Activity'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Activities Leading to Shark Attacks')
plt.xlabel('Activity')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

*This bar chart shows the most common activities people were engaged in when shark attacks occurred, identifying activities with higher risk.*

### 3. Top 10 Areas with Shark Attacks in the USA

```python
usa_attacks = df[df['Country'] == 'USA']

plt.figure(figsize=(12, 6))
usa_attacks['Area'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Areas with Shark Attacks in the USA')
plt.xlabel('Area (State/Region)')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

*Focusing on the USA, this bar chart highlights the specific areas (e.g., states or regions) within the country that experience the most shark attacks.*

### 4. Shark Attacks by Time of Day

```python

df['Time_of_Day'] = df['Hour'].apply(get_time_of_day)

plt.figure(figsize=(10, 6))
sns.countplot(x='Time_of_Day', data=df, order=['Morning', 'Afternoon', 'Evening', 'Night'])
plt.title('Shark Attacks by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Attacks')
plt.show()
```

*This count plot analyzes the frequency of shark attacks based on the time of day, revealing if certain periods are more prone to incidents.*

### 5. Top 10 Shark Species Involved in Attacks

```python
# Print all columns to debug the KeyError
print("DataFrame columns:", df.columns.tolist())

plt.figure(figsize=(12, 6))
# Check if 'Species ' column exists before attempting to plot
if 'Species ' in df.columns:
    df['Species '].value_counts().head(10).plot(kind='bar')
    plt.title('Top 10 Shark Species Involved in Attacks')
    plt.xlabel('Shark Species')
    plt.ylabel('Number of Attacks')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Species ' not found in DataFrame. Please check the column names printed above.")
```

*This visualization highlights the most frequently involved shark species in recorded attacks. It helps identify which species are more commonly associated with human encounters. Typically, species like great white sharks, tiger sharks, and bull sharks appear at the top due to their size, habitat, and proximity to coastal areas.*

### 6.Age Distribution of Victims

```python
plt.figure()
sns.histplot(df['Age_numeric'].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Victims")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

*The age distribution of shark attack victims shows how incidents are spread across different age groups. The visualization typically reveals that young adults (around 15â35 years) are most frequently affected, likely due to higher participation in water activities such as swimming and surfing.*

### 7.Gender-wise Shark Attacks

```python
plt.figure()
sns.countplot(x='Sex ', data=df)
plt.title("Gender-wise Shark Attacks")
plt.show()
```

*This visualization shows the distribution of shark attacks between males and females. It typically reveals that males are involved in a significantly higher number of attacks compared to females.*

### 8.Correlation Heatmap


```python
# Convert 'Age' to numeric, coercing errors to NaN
df['Age_numeric'] = pd.to_numeric(df['Age'], errors='coerce')

plt.figure() # Calculate correlation using the new numeric 'Age_numeric' column
corr = df[['Year', 'Age_numeric']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

*A correlation heatmap visualizes the relationship between numerical variables in the dataset using color intensity. It helps identify whether variables are positively correlated (move together), negatively correlated (move oppositely), or have little to no relationship.*

### 9.Type of Shark Attacks by Top 10 Countries

```python
# Get the top 10 countries with the most shark attacks
top_10_countries = df['Country'].value_counts().head(10).index
# Filter the DataFrame to include only these top countries
df_top_countries = df[df['Country'].isin(top_10_countries)]
# Create a countplot for 'Type' of attack by 'Country'
plt.figure(figsize=(15, 8))
sns.countplot(data=df_top_countries, x='Country', hue='Type', palette='viridis', order=top_10_countries)
plt.title('Type of Shark Attacks by Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

*This visualization shows the distribution of different types of shark attacks (e.g., fatal vs non-fatal) across the top 10 countries with the highest number of incidents. It helps compare how the nature and severity of attacks vary by region.*

### 10. Shark Attacks by Activity by gender

```python
# Get the top 10 activities with the most shark attacks
top_10_activities = df['Activity'].value_counts().head(10).index

# Filter the DataFrame to include only these top activities
df_top_activities = df[df['Activity'].isin(top_10_activities)].copy()

# Ensure 'Sex ' column is cleaned (remove leading/trailing spaces and standardize)
df_top_activities.loc[:, 'Sex '] = df_top_activities['Sex '].str.strip().str.upper()
df_top_activities.loc[:, 'Sex '] = df_top_activities['Sex '].replace({'M': 'Male', 'F': 'Female', 'N': 'Unknown', '.': 'Unknown', 'UNOC': 'Unknown', 'LLI': 'Unknown', '2017': 'Unknown'})

# Create a countplot for 'Activity' by 'Sex '
plt.figure(figsize=(15, 8))
sns.countplot(data=df_top_activities, x='Activity', hue='Sex ', palette='deep', order=top_10_activities)
plt.title('Shark Attacks by Top 10 Activities and Sex')
plt.xlabel('Activity')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sex', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

*This visualization shows how shark attacks are distributed across different activities (such as surfing, swimming, fishing, etc.) for each gender. It helps identify which activities are most commonly associated with attacks among males and females.*
